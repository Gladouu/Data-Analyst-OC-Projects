### Question 3 

**Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.**

Vous ferez cela à partir de ces informations :
Population de chaque pays ;
Disponibilité alimentaire donnée pour chaque produit et pour chaque pays en kcal/personne/jour.
Disponibilité alimentaire en protéines donnée pour chaque produit et pour chaque pays en g/personne/jour.




```python
# Je renomme les colonnes de mon DF animaux et j'ajoute un colonne 'origine' respectivement à mon DF vgt et 
# animaux.

animaux.columns = ['codeDomaine', 'domaine', 'codePays', 'pays', 'codeElement', 'element', 'codeProduit',
                   'produit', 'codeAnnee', 'annee', 'unite', 'valeur', 'symbole', 'descriptionSymbole']

animaux['origine'] = 'animaux'
vgt['origine'] = 'vegetaux'

# Comme dans le fichier population la Chine était representée plusieurs fois, je décide de vérifier si c'est 
# aussi le cas dans les fichiers animaux et végétaux :

animaux[animaux['pays'].str.contains('Chine')]
```

```python
# La Chine est donc également représentée plusieurs fois dans le fichier animaux. Je procède de même avec le
# fichier végétaux. 

# De ce fait : 

animaux = animaux[~animaux.codePays.isin([96, 128, 41, 214])]
animaux.reset_index() 
vgt = vgt[~vgt.codePays.isin([96, 128, 41, 214])]
vgt.reset_index().head(2)
```

```python
# L'information sur la disponnibilité alimentaire en kcal et en kg de protéines étant à la fois présente dans 
# le DF animaux et dans le DF végétaux, je décide donc de regrouper les deux DF pour avoir cette information à
# un seul endroit. 

vgt_ani = [vgt, animaux]
vgt_ani = pd.concat(vgt_ani, ignore_index = True)
```

```python
vgt_ani = vgt_ani.loc[:, ['codePays', 'pays', 'element', 'codeProduit', 'produit', 'annee',  'unite', 'valeur',
                      'origine', ]]
```

```python
# J'utilise ensuite une table pivot pour avoir mes éléments en colonne afin que cela soit plus simple pour la
# suite.

vgt_ani = pd.pivot_table(vgt_ani, 
                         index = ['codePays', 'pays', 'annee', 'origine', 'codeProduit', 'produit'], 
                         columns = ['element'],
                         values = ['valeur'], 
                         aggfunc = sum).reset_index()


# On prendra le soin de noter ici qu'hormis pour la disponibilité alimentaire (Kcal/personne/jour), la disponi-
# bilité alimentaire en quantité (kg/personne/an), la disponibilité de matière grasse en quantité (g/personne/
# jour), et la disponibilité de protéines en quantité (g/personne/jour), le reste des éléments est exprimé en 
# milliers de tonnes.
```

```python
vgt_ani.head()
```

```python
vgt_ani.columns.values
```

```python
vgt_ani.columns = ['codePays', 'pays', 'annee', 'origine', 'codeProduit', 'produit', 'alimAnim', 
                               'autres', 'dispoKcalPJ','dispoKgPA', 'dispoGrasgPJ', 'dispoProtgPJ',
                               'dispoInt', 'exp', 'imp', 'nourriture', 'pertes', 'production', 'semences', 
                               'traitement', 'var']

pop['valeur'] *= 1000
pop_filter = pop.loc[:, ('codePays', 'pays', 'valeur')]

# J'effectue une jointure entre vgt_ani et pop_filter 

join = pd.merge(vgt_ani, pop_filter, on = ['pays', 'codePays'])
join
```

```python
join = join.rename({'valeur' : 'population'}, axis = 1)

# Pour calculer la disponibilité alimentaire en Kcal (sur l'année 2013) je rajoute donc une colonne à join qui 
# n'est autre que la multiplication de la colone dispoKcalPJ par la population et par 365.

# Je fais de même pour la disponibilité alimentaire en protéines, en prenant le soin de diviser la valeur de la
# colonne dispoProtgPJ par 1000 pour obtenir des kg et non des g (= CONV_KG_G). 

#Conversion de kg en g : 

CONV_KG_G = 1000


join['dispoKcal'] = join['dispoKcalPJ'] * 365 * join['population']

join['dispoProtKg'] = (join['dispoProtgPJ'] / CONV_KG_G) * 365 * join['population']

joinQ3filter = join.loc[:, ('codePays','pays', 'annee', 'origine', 'codeProduit', 'produit', 'dispoKcalPJ', 
                           'dispoProtgPJ', 'dispoKcal', 'dispoProtKg')]

joinQ3filter.head()
```

### Question 4 

**A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire (pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg. Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio aux données disponibles sur internet, par exemple en cherchant la valeur calorique d'un oeuf.**

Indication : La disponibilité alimentaire en kcal/personne/jour est calculée par la FAO en multipliant la quantité Nourriture par le ratio énergie/poids (en kcal/kg), puis en le divisant par la population du pays puis par 365. Ici, on vous demande juste de retrouver le ratio énergie/poids que la FAO a utilisé dans son calcul.

**En suivant la même méthodologie, calculez également le pourcentage de protéines de chaque produit (pour chaque pays). Ce pourcentage est obtenu en calculant le ratio "poids de protéines/poids total" (attention aux unités utilisées). Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio aux données disponibles sur internet, par exemple en cherchant la teneur en protéines de l'avoine.**

```python
# Si la disponibilité alimentaire en kcal est égale à la nourriture multipliée par le ratio "energie/poids" 
# alors le ratio "énergie/poids" est égal à la disponibilité alimentaire en kcal divisée par la nourriture.

# Pour rappel, cette dernière étant en milliers de tonnes, il faudra multiplier sa valeur par 10^6 (= CONV_KT_ 
# KG).

# Conversion milliers de tonnes en kg :

CONV_KT_KG = 10**6 

join['ratio'] = join['dispoKcal'] / (join['nourriture'] * CONV_KT_KG)

joinQ4filter = join.loc[:, ('codePays','pays', 'annee', 'origine', 'codeProduit', 'produit', 'dispoKcalPJ', 
                           'dispoProtgPJ', 'dispoKcal', 'dispoProtKg', 'ratio')]

joinQ4filter.head()
```

```python
# Je vérifie la cohérence de mon calcul : 

joinQ4filter[joinQ4filter.produit.isin(['Oeufs']) & join.pays.isin(['France'])]
```

[Valeur calorique d'un oeuf](https://fr.wikipedia.org/wiki/Œuf_(aliment))

```python
# D'après la ressource ci-dessus, la valeur calorique d'un oeuf pour 100 g est de 147 kcal.

# Notre ratio étant en kcal par kg (et non par g), si l'on divise 147 par 0,1 pour avoir ce ratio par kg (100g
# = 0,1 kg) on obtient 1470, ce qui est proche de la valeur retrouvée du ratio pour l'oeuf en France. 
```

En suivant la même méthodologie, calculez également le pourcentage de protéines de chaque produit (pour chaque pays). Ce pourcentage est obtenu en calculant le ratio "poids de protéines/poids total" (attention aux unités utilisées). Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio aux données disponibles sur internet, par exemple en cherchant la teneur en protéines de l'avoine.

```python
# Je divise par 100 pour avoir un pourcentage

join['pourcentProt'] = join['dispoProtKg'] / ((join['nourriture'] * CONV_KT_KG)/100)
join.head()
```

```python
# Je vérifie la cohérence de mon calcul : 

avoine = join[join.produit.isin(['Avoine'])]

avoine = avoine[(avoine['pourcentProt'] > 0)   
                & (avoine['pourcentProt'] != np.inf)]

print("Le pourcentage moyen de teneur en protéine pour l'avoine est de : " + 
      str(round(avoine['pourcentProt'].mean(),2)) + " %.")
```

**En Thaïlande, quelle proportion de manioc est exportée ? Quelle est la proportion de personnes en sous-nutrition?**
